```python
#Import libraries
import altair as alt
import pandas as pd
```

```python
df = pd.read_csv('data_master.csv')
df = df[['year', 'acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'speechiness', 'valence']]
df
```

```python
data = df.groupby('year').mean().reset_index()
data = data.melt('year').rename(columns={'year': 'Year', 'value': 'Value', 'variable': 'Attribute'})
data
```

```python
selection = alt.selection_multi(fields=['Attribute'], bind='legend')

alt.Chart(data).mark_line(
).encode(
    x='Year',
    y='Value',
    color='Attribute',
    opacity=alt.condition(selection, alt.value(1), alt.value(0.2))
).add_selection(
    selection
).properties(
    width=600, height=300
)
```
